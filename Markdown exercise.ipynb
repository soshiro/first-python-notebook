# Import pandas as pd

```python
import pandas as pd
```

# Read in proposition committees

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

# Read in contributions

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

# Find full Proposition 66 name

```python
props.prop_name.value_counts()
```

# Find committees for/against Prop 66

```python
props[props.prop_name == 'PROPOSITION 066- DEATH PENALTY. PROCEDURES. INITIATIVE STATUTE.']
```

# Save Prop 66 committees as deathcom

```python
deathcom = props[props.prop_name == 'PROPOSITION 066- DEATH PENALTY. PROCEDURES. INITIATIVE STATUTE.']
```

# Inspect deathcom

```python
deathcom.head()
```

```python
deathcom.info()
```

# Merge Prop 66 committees with contributions

```python
pd.merge(deathcom, contribs, on="calaccess_committee_id")
```

# Save Prop 66 committees + contributions as merged

```python
merged = pd.merge(deathcom, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

# Sum Prop 66 contributions

```python
merged.amount.sum()
```

# How many contributions support/oppose

```python
merged.committee_position.value_counts()
```

# Total amount support/oppose

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

# Percentage support/oppose

```python
support.amount.sum() / merged.amount.sum()
```

```python
oppose.amount.sum() / merged.amount.sum()
```

# Contributions in descending order

```python
merged.sort_values("amount", ascending=False)
```
